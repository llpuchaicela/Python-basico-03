```python
# Generar un archivo .csv que contenga únicamente las características:
# Nombre_Institucion
# Provincia 
# Canton
# Parroaquia
# Número de estudiantes
# Número de docentes

result = datos[["Nombre_Institucion", "Provincia", "Canton", "Parroquia", "Numero_de_estudiantes", "Numero_de_docentes"]]

# Siempre que cumpla con la condición: el número de estudiantes debe ser mayor a 20 y menor o igual a 100
result = result[(result["Numero_de_estudiantes"]>20) & (result["Numero_de_estudiantes"]<=100) ]

# El archivo debe tener el nombre: instituciones_0001.csv
result.to_csv("data/instituciones_0001.csv")
```
